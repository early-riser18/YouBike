```python
import pandas as pd
```

```python
df_ytd = pd.read_csv('raw_data/weather_data_raw_2024-02-19_12:39:08.csv')
df_tod = pd.read_csv('raw_data/weather_data_raw_2024-02-20_11:26:22.csv')
```

```python
df_ytd
```

# Compare change in future predictions

```python
# Yesterday predictions - anything after 1 PM
df_ytd['datetime'] =  pd.to_datetime(df_ytd['datetime']).dt.tz_convert(tz='Asia/Taipei')
dt_start_prediction = pd.to_datetime('2024-02-20 11:00:00', format='%Y-%m-%d %H:%M:%S').tz_localize(tz='Asia/Taipei')
ytd_pred = df_ytd[df_ytd['datetime'] > dt_start_prediction]

```

```python
#today pred - anything after 11am
df_tod['datetime'] = pd.to_datetime(df_tod['datetime']).dt.tz_convert(tz='Asia/Taipei')
dt_start_prediction_Tod = pd.to_datetime('2024-02-20 11:00:00', format='%Y-%m-%d %H:%M:%S').tz_localize(tz='Asia/Taipei')
tod_pred = df_tod[df_tod['datetime'] > dt_start_prediction_Tod]
```

```python
pred_comp = pd.merge(
    left=ytd_pred,
    right=tod_pred,
    how='left',
    on=['lat', 'datetime'],
    suffixes=['_ytd', '_tod']
)
```

```python
pred_comp[['temperature_ytd', 'temperature_tod', 'datetime']]
```

```python
pred_comp['pred_temp_diff'] = pred_comp['temperature_ytd'] - pred_comp['temperature_tod']
pred_comp['pred_temp_diff'].describe()
```
